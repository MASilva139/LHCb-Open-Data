# Más información sobre la violación CP


<div align="justify">¿Por qué el universo contiene mucha más materia que antimateria? Esta es una de las principales preguntas pendientes de la física fundamental. Se espera que el big-bang haya producido cantidades iguales de materia y antimateria. El universo actual está hecho de materia.

<img src="../Images/Mirror.png" width="600" />

Una diferencia fundamental entre el comportamiento de la materia y la antimateria (violación CP) es uno de los requisitos, conocidos como condiciones de Sájarov, para explicar cómo esto pudo haber surgido. [Puedes leer el artículo original de Sájarov aquí](https://inspirehep.net/record/51345?ln=en).</div>

La simetría CP es el producto de dos simetrías: C para la conjugación de carga, que transforma una partícula en su antipartícula, y P para la paridad, que refleja las coordenadas espaciales. [Una violación de esta simetría fue descubierta en 1964](http://journals.aps.org/prl/pdf/10.1103/PhysRevLett.13.138) en desintegraciones a través de la fuerza débil de partículas llamadas kaones, por lo que se concedió el [Premio Nobel de 1980](http://www.nobelprize.org/nobel_prizes/physics/laureates/1980/). Un mecanismo para incluir la violación CP en el [Modelo Estándar](https://en.wikipedia.org/wiki/Standard_Model) de la física de partículas fue sugerido en 1973 y verificado mediante pruebas de violación CP en desintegraciones débiles de los mesones B en los [experimentos BaBar y Belle](http://paperity.org/p/52588259/the-physics-of-the-b-factories) desde 1999 hasta 2010, concediéndose el [Premio Nobel de 2008](http://www.nobelprize.org/nobel_prizes/physics/laureates/2008/) por este trabajo. Se puede encontrar más información sobre la violación CP en muchos libros de texto de física de partículas de nivel universitario, como <a href="http://web-docs.gsi.de/~gkalicy/books/Martin%20-%20Particle%20Physics%203e%20(Wiley,%202008).pdf">Particle Physics</a> de B.R. Martin y G. Shaw, o en una serie de <a href="http://www.hep.manchester.ac.uk/u/parkes/Teaching/CPViolation/PP.html">conferencias sobre violación CP</a> de Chris Parkes en la Universidad de Manchester. 

Los experimentos del LHC han descubierto la última partícula del [Modelo Estándar](http://home.cern/about/physics/standard-model), el [bosón de Higgs](http://home.cern/topics/higgs-boson), y están dedicados a buscar nueva física más allá del Modelo Estándar. El Modelo Estándar (SM) de la física de partículas incluye la violación CP, pero la magnitud del efecto es demasiado pequeña para explicar el desequilibrio materia-antimateria en el universo. En consecuencia, se requieren fuentes adicionales de violación CP. Nuevas partículas más allá del SM, en general, añadirán fuentes adicionales de violación CP. Por lo tanto, las mediciones de la violación CP proporcionan un método importante para buscar nuevos fenómenos más allá del SM. Estas partículas de nueva física también podrían potencialmente proporcionar explicaciones para otras grandes preguntas sin respuesta, como la naturaleza de la [materia oscura](https://en.wikipedia.org/wiki/Dark_matter).

Los estudios experimentales de la violación CP se pueden realizar comparando las desintegraciones de un estado de partícula y antipartícula. El experimento LHCb es el experimento del LHC dedicado a tales estudios y ha producido una serie de resultados importantes. Los aspectos más destacados de su programa de física, actualizados regularmente, se describen en esta [página web](http://lhcb-public.web.cern.ch/lhcb-public/). Las publicaciones científicas completas de la colaboración están [disponibles gratuitamente](http://lhcbproject.web.cern.ch/lhcbproject/Publications/LHCbProjectPublic/Summary_all.html).

El análisis realizado en este proyecto se basa en los efectos de violación CP más grandes que se han observado. Los artículos están disponibles [aquí](http://journals.aps.org/prl/pdf/10.1103/PhysRevLett.112.011801) y [aquí](http://journals.aps.org/prl/pdf/10.1103/PhysRevLett.111.101801). La muestra de datos proporcionada en este proyecto es la misma que se utilizó para estas publicaciones. Se está realizando más investigación sobre el origen de estos efectos de violación CP y hay un artículo adicional, que utiliza un conjunto de datos más grande y más análisis, [aquí](https://cds.cern.ch/record/1751517).
