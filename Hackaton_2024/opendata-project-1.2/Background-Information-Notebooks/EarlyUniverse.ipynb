# Física de Partículas y el Universo Temprano

En los albores del tiempo, el universo atravesó eras en las que consistía únicamente en partículas elementales. En la siguiente imagen, puedes ver esto dentro de la Era GUT (Gran Unificación), la Era Electro débil y la Era de Partículas.

<img src="../Images/bigbang_timeline.jpg" width="700" />

Estas partículas luego se combinaron para formar núcleos y átomos, y estos se agruparon para formar las estrellas y galaxias que vemos ahora, 13.8 mil millones de años después. Los experimentos en colisionadores de partículas, como el Gran Colisionador de Hadrones, nos permiten recrear condiciones similares a las del universo temprano y estudiar el comportamiento de las partículas elementales. Esto nos permite estudiar el comportamiento de las partículas y antipartículas en el universo temprano que condujo al universo dominado por la materia.
