# Formato de los Datos de Evento

Los archivos root que se te han proporcionado contienen información sobre muchos eventos diferentes. Para cada evento, contienen propiedades que se han medido para ese evento y que pueden ser útiles en tu análisis.

En cada evento se han identificado tres partículas que son candidatas a ser partículas producidas en la desintegración de los mesones B. La información de estas tres partículas está etiquetada como H1_xx, H2_xx, H3_xx. Aquí 'xx' es la propiedad particular de la partícula. H1_PX es la componente X del momento de la partícula 1. Las variables correspondientes para las partículas 2 y 3 se llaman H2_PX y H3_PX.

Las variables disponibles en este análisis se enumeran en la siguiente tabla.

Para los datos reales, todas las variables están completadas para cada evento. Para los datos simulados, no todas las variables están completadas.

No todas las variables listadas en la tabla son necesarias para completar este proyecto. Se han incluido variables adicionales que fueron importantes en la selección de la muestra de datos y que podrían ser útiles para extensiones del proyecto.

Para aquellos que han estudiado Física formalmente, algunas de estas variables les serán familiares, pero es probable que otras no lo sean.

Las variables que necesitarás para este proyecto de análisis se explican más detalladamente aquí:

<ul> 
<li> <strong>H1_PX:</strong> La componente medida del momento de la primera partícula (H1) en la dirección x. Reemplaza X con Y y Z para las otras variables. Usarás los momentos al calcular energías y masas.</li>
<li> <strong>H1_ProbK:</strong> Una estimación de la probabilidad de que la primera partícula sea un kaón. En este análisis buscas analizar las desintegraciones de mesones B en tres kaones. Por lo tanto, querrás seleccionar eventos donde las probabilidades de que las tres partículas sean kaones sean altas.</li>
<li> <strong>H1_ProbPi:</strong> Una estimación de la probabilidad de que la primera partícula sea un pion. Dado que en tu análisis las partículas del estado final no deberían ser piones, querrás seleccionar eventos donde las probabilidades de que las tres partículas sean piones sean bajas.</li>
<li> <strong>H1_Charge:</strong> La carga eléctrica (en unidades de la carga del electrón) de la primera partícula.</li>
<li> <strong>H1_isMuon:</strong> Un valor binario que indica si es probable que la partícula sea un muón. Como deseas que tus partículas sean kaones, puedes verificar que esto sea falso (0).</li> 
</ul>
Las variables adicionales que se proporcionan son:

<ul> 
<li> <strong>B_FlightDistance:</strong> El punto de colisión de los protones se encuentra a partir de las trayectorias reconstruidas que provienen de ese punto. El vértice de desintegración del mesón B se reconstruye buscando el punto donde se encuentran las tres trayectorias del estado final. La distancia entre el punto de colisión, donde se crea el mesón B, y el punto de desintegración del mesón B, donde se crean las tres partículas del estado final, es la distancia de vuelo del mesón B. El mesón B es una partícula de vida corta y típicamente viajará unos pocos mm hasta un cm antes de desintegrarse.</li>

<li> <strong>B_VertexChi2:</strong> Puedes encontrar el punto de desintegración (vértice) del mesón B encontrando el punto donde se encuentran estas tres trayectorias del estado final. En la práctica, las trayectorias medidas no provendrán de un punto exacto debido a los errores de medición, pero se encuentra la mejor estimación del punto. El χ<sup>2</sup> (Chi2 o <a href="https://en.wikipedia.org/wiki/Chi-squared_distribution">distribución chi-cuadrado</a>) es una medida estadística de la compatibilidad de las tres trayectorias de haber provenido de un punto común. Un valor grande (mayor que 1) sugiere que las trayectorias realmente no provienen de un punto común y, por lo tanto, es más probable que no sean de una desintegración real de un mesón B.</li>

<li> <strong>H1_IPChi2:</strong> El parámetro de impacto se define como la distancia perpendicular entre la trayectoria de la partícula y el vértice de colisión protón-protón. Las tres trayectorias del estado final provienen del punto de desintegración del mesón B y, por lo tanto, no deberían apuntar de vuelta al vértice de colisión. La mayoría de las trayectorias en un evento se producen en el punto de colisión, por lo que si una trayectoria apunta de vuelta al vértice de colisión, es una indicación de que podría no provenir de la desintegración del mesón B. El χ<sup>2</sup> es una medida de la compatibilidad de esta trayectoria de provenir del vértice de colisión, teniendo en cuenta las incertidumbres.</li>
</ul>
