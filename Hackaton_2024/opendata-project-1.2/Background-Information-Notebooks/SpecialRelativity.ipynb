# Relatividad Especial y Masa Invariante

<img src="https://upload.wikimedia.org/wikipedia/commons/5/50/Albert_Einstein_(Nobel).png" width="200" />

La relatividad especial describe la relación entre el tiempo y el espacio. Se basa en dos postulados.

**Primer Postulado de Einstein**: Las leyes de la física son las mismas en todo sistema de referencia inercial.

**Segundo Postulado de Einstein**: La velocidad de la luz en el vacío es la misma en todos los sistemas de referencia inerciales y es independiente del movimiento de la fuente.

Tiene una serie de consecuencias, incluyendo la contracción de la longitud, la dilatación del tiempo y el límite universal de velocidad de la luz. El elemento que nos concierne para este proyecto es que propone una equivalencia entre masa y energía.

<div align="justify">La equivalencia masa-energía se expresa mediante la famosa fórmula E=mc<sup>2</sup>, donde c es la velocidad de la luz en el vacío. Esto da la energía en reposo, E, de un cuerpo de masa en reposo, o masa invariante, m. La energía en reposo y la masa en reposo son la energía y la masa que tiene el cuerpo cuando no está en movimiento.

La fórmula completa incluye el componente de energía cinética de la energía, es decir, la energía que tiene el cuerpo debido a su movimiento. Esta es la versión que necesitamos. Relaciona la energía, E, de una partícula; la masa invariante, m, de la partícula; y la magnitud del momento, p, de la partícula. El momento es una cantidad vectorial y aquí nos interesa la magnitud del vector. La expresión es

<center> $E^2 = (pc)^2 + (mc^2)^2$ </center>

En este proyecto trabajamos en [unidades naturales](https://en.wikipedia.org/wiki/Natural_units). Expresamos las energías en MeV, los momentos en MeV/c y las masas en MeV/c<sup>2</sup>. No tenemos que preocuparnos por los factores 'c', y podemos trabajar con la versión simplificada de la ecuación

<center> $E^2 = p^2 + m^2$ </center>
