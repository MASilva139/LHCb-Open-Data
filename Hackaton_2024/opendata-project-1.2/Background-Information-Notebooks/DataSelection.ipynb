# Selección de datos

La muestra de datos que se te proporciona tiene aplicados cortes de preselección para seleccionar una muestra de eventos que es mayoritariamente los eventos que nos interesa analizar, conocidos como "señal", con un número relativamente bajo de eventos de otros procesos de "fondo". Los cortes de preselección más importantes que se han aplicado se enumeran en la tabla siguiente y se explican aquí:
<ul>
<li><strong>Momento.</strong> Las tres trayectorias del estado final deben tener todas un momento significativo (p) y un momento en el plano transversal a la dirección del haz (p<sup>T</sup>), ya que provienen de la desintegración de un mesón B que tiene una masa relativamente alta, mientras que las trayectorias de procesos de fondo típicamente tendrán momentos bajos.</li> <li><strong>Masa del B.</strong> La masa del mesón B candidato se puede calcular a partir de los momentos medidos de las partículas del estado final (bajo el supuesto de si son kaones o piones) y se seleccionan eventos con masas cercanas a la masa conocida del mesón B.</li> <li><strong>Parámetros de impacto.</strong> El mesón B tiene una vida media de 1.6 × 10<sup>12</sup> s y típicamente viajará una distancia de unos pocos mm en LHCb. La distancia entre su punto de producción, donde colisionan los haces de protones, conocido como vértice primario, y el vértice de desintegración en las tres trayectorias del estado final es clave para identificar una muestra limpia de eventos. La distancia mínima de aproximación de las trayectorias del estado final al vértice primario se conoce como parámetro de impacto (IP). Estas trayectorias tendrán, por tanto, un IP significativo. El IP χ<sup>2</sup> es una medida de si la trayectoria es compatible con haber provenido del vértice primario; una trayectoria de un proceso de fondo que se origina en el vértice primario típicamente tendrá un IP χ<sup>2</sup> bajo.</li>
<p>
La muestra no ha tenido ninguna identificación de las partículas del estado final aplicada y, por lo tanto, contiene todas las combinaciones de piones y kaones cargados. La selección de eventos es común a todos los canales donde un mesón B se desintegra en tres trayectorias en el estado final, ya que la cinemática de todos los canales será muy similar. Utilizando la información de los sistemas de identificación de partículas del detector, particularmente los detectores RICH para separar piones y kaones, se puede separar el canal de desintegración que se va a estudiar. Este es un proceso estadístico; todavía habrá una tasa significativa de identificación errónea de la partícula del estado final como pión/kaón para mantener una alta eficiencia en la selección de la señal.
