# Enfoque de Física de Partículas de este Proyecto

En este proyecto analizarás partículas que contienen el quark *b* (o beauty) y sus antipartículas. Para obtener más información sobre ellos, puedes leer los enlaces de información general sobre el [Modelo Estándar](http://home.cern/about/physics/standard-model).

Estos quarks *b* eran comunes poco después del Big Bang, pero no son un componente de la materia que nos rodea en el universo actual. Los colisionadores de partículas como el LHC se utilizan para producir quarks *b*, así como sus contrapartes de antimateria, los quarks anti-b.

Los quarks no se observan de forma aislada; se combinan en partículas compuestas. Las combinaciones de tres quarks se conocen como bariones: los protones y neutrones son los bariones más comunes. De manera similar, los antiquarks correspondientes pueden combinarse en sus equivalentes de antimateria: antiprotones y antineutrones. Las combinaciones de un quark y un antiquark se conocen como mesones: los piones y kaones son los mesones más comunes. Todas las partículas compuestas por quarks y antiquarks (bariones y mesones) se conocen como hadrones.

En este proyecto estudiaremos las desintegraciones de los mesones B<sup>+</sup> y B<sup>-</sup>. Un B<sup>+</sup> tiene una carga eléctrica de +1 unidad y está compuesto por un quark *u* (up) y un quark *anti-b* (beauty). A B<sup>-</sup> is the anti-particle of the B<sup>+</sup>. Un B<sup>-</sup> es la antipartícula del B<sup>+</sup>. Tiene una carga eléctrica de -1 unidad y está compuesto por un quark anti-u (up) y un quark *b* (beauty).

Los mesones B<sup>+</sup> y B<sup>-</sup> tienen una vida media corta (10<sup>-12</sup> s) y se desintegran a través de la fuerza débil en otras partículas. En este análisis estudiaremos las desintegraciones de los mesones B<sup>+</sup> y B<sup>-</sup> en tres kaones cargados (K<sup>+</sup> o K<sup>-</sup>). El K<sup>+</sup> es un mesón compuesto por un quark *u* y un quark *anti-s* (strange), y tiene una carga de +1 unidad. El K<sup>-</sup> es su antipartícula.

Los mesones B<sup>+</sup> y B<sup>-</sup> no se observan directamente en el detector debido a su corta vida útil. Incluso viajando cerca de la velocidad de la luz, y teniendo en cuenta los efectos de la relatividad, solo recorren unos pocos mm en el detector antes de desintegrarse. Los kaones cargados tienen vidas lo suficientemente largas (10<sup>-8</sup> s) como para atravesar el detector LHCb. Medimos las propiedades de estos kaones en el detector, por ejemplo determinando su momento e identificándolos, y son las mediciones de estos kaones las que usaremos para el análisis.

En este proyecto estudiarás las desintegraciones
<p>
B<sup>+</sup> → K<sup>+</sup>K<sup>+</sup>K<sup>−</sup>

y el equivalente de antipartícula

<p>
B<sup>-</sup> → K<sup>+</sup>K<sup>-</sup>K<sup>−</sup>.


## Diagramas de partículas

Los físicos de partículas utilizan diagramas para representar las desintegraciones de las partículas. Estos se conocen como [diagramas de Feynman](https://en.wikipedia.org/wiki/Feynman_diagram) o diagramas de flujo de quarks. Aquí hay un ejemplo de diagrama para la desintegración que estudiarás.

<img src="../Images/FeynDiag.png" width="600" />

La imagen superior muestra los quarks como líneas en el diagrama (*u*, *s*, *b*). Los antiquarks se indican con una barra sobre ellos. El tiempo corre de izquierda a derecha en el diagrama. Los corchetes muestran cómo las partículas se combinan en mesones. El proceso comienza con un B<sup>+</sup> en el lado izquierdo. Este se desintegra en los tres kaones cargados, dos K<sup>+</sup> y un K<sup>-</sup> en el lado derecho. El W<sup>+</sup> es el portador de la fuerza débil; el quark anti-*b* se está desintegrando a través de la fuerza débil en este proceso.

Esta imagen muestra solo la desintegración del mesón B positivo. La desintegración del mesón B negativo se vería igual pero con cada partícula reemplazada por su antipartícula: cada quark sería reemplazado por su antiquark y viceversa, y los signos '+' se reemplazan por signos '-' y viceversa.

## Proyecto

Comenzarás reconstruyendo el momento, la energía y la masa del mesón B utilizando los momentos medidos de los kaones. Luego buscarás diferencias en las tasas a las que ocurren el proceso de desintegración del B<sup>+</sup> y el proceso de desintegración del B<sup>-</sup>: esto sería una diferencia en el comportamiento de la materia y la antimateria (violación CP). Luego separarás las desintegraciones por sus características en regiones 'locales' e identificarás partículas intermedias que se producen durante las desintegraciones. Finalmente, harás una observación clara de las diferencias entre materia y antimateria observando las diferencias entre las regiones locales para las desintegraciones B<sup>+</sup> y B<sup>-</sup>. ¡Si logras trabajar a lo largo de todo el proyecto hasta el final, observarás la violación CP por ti mismo!
