# Diagramas de Dalitz

## ¿Qué son los diagramas de Dalitz?

Un diagrama de Dalitz es una representación visual del espacio de fases de una desintegración de tres cuerpos. Lleva el nombre de su inventor, Richard Dalitz (1925–2006).

Si no hay resonancias intermedias en la desintegración de tres cuerpos que se está analizando, el diagrama está poblado uniformemente con eventos dentro de sus límites físicos.

Si hay resonancias intermedias, estas aparecen como bandas en el diagrama con una mayor densidad de eventos. Tomemos el caso donde B<sup>+</sup> → K<sup>+</sup>R<sup>0</sup>, donde R<sup>0</sup> es una resonancia de partícula neutra que puede desintegrarse R<sup>0</sup> → K<sup>+</sup>K<sup>-</sup>. Observaríamos una banda mejorada de eventos a lo largo del diagrama de Dalitz en una masa correspondiente a la de la partícula R<sup>0</sup>. Las partículas también tienen una propiedad de la mecánica cuántica llamada espín. Si R<sup>0</sup> tiene espín 0, entonces observaremos una banda continua de eventos en todo el diagrama. Si R<sup>0</sup> tiene espín 1, entonces la banda tendrá un mínimo; para partículas de espín 2, tendrá dos mínimos. Cuando las resonancias se superponen, pueden interferir entre sí, una propiedad análoga a la interferencia en la luz con la que quizás estés familiarizado.

## ¿Cómo construimos un diagrama de Dalitz?

En cada uno de los dos ejes del diagrama de Dalitz trazamos una cantidad con unidades de masa al cuadrado. Cada una de estas masas corresponde a las masas de una cuasipartícula (R<sup>0</sup> arriba). Es la masa invariante de dos partículas del estado final combinadas. Si llamamos a las partículas del estado final 1, 2, 3, estas masas se pueden escribir como m<sub>12</sub>, m<sub>23</sub>. El cuadrado de estas masas es la masa trazada en los ejes, como en este diagrama (crédito de la imagen: particle data group).

Estas masas están restringidas en los valores que pueden tomar; hay masas máximas y mínimas posibles que definen los límites del diagrama, ya que la energía y el momento se conservan.
