# Información sobre el Detector, el Software y la Muestra de Datos

El detector del experimento LHCb es un detector de un solo brazo orientado hacia adelante. Este detector se especializa en el estudio de las diferencias entre materia y antimateria mediante el análisis de procesos de desintegración de quarks *b* (beauty) o *c* (charm) quark (y anti-quark).

La fotografía muestra el detector real en su caverna en el LHC, aproximadamente a 100 metros bajo tierra. La imagen ha sido reflejada para tener la misma orientación que el diagrama de disposición.

El sistema de coordenadas utilizado es tal que el eje z describe la trayectoria de uno de los haces de protones hacia el interior del detector (de izquierda a derecha). El eje y es vertical y el eje x es perpendicular a la página.

A las energías del LHC, los quarks *b* (beauty) o *c* (charm) se producen con un fuerte impulso hacia adelante a lo largo de las direcciones de los haces de protones, y aquellos que vuelan a través del detector son medidos. Los mesones B, que contienen el quark *b*, se desintegran a unos pocos mm (hasta un cm) dentro del detector. Los productos de desintegración (en este proyecto K<sup>+</sup> y K<sup>-</sup>) se miden en el detector y se reconstruye su trayectoria a través de él. Combinando estos productos de desintegración podemos reconstruir el mesón B.

El detector incluye un sistema de seguimiento de alta precisión que detecta partículas cargadas. Un detector de vértice de tiras de silicio, conocido como localizador de vértices ([VELO](http://lhcb-public.web.cern.ch/lhcb-public/en/detector/VELO-en.html)), rodea la región de interacción pp y mide al menos tres (típicamente diez) puntos en cada trayectoria. Esto permite reconstruir la posición del punto de interacción pp (vértice primario) y medir la posición de la desintegración del hadrón que contiene el quark b (vértice secundario). Las partículas cargadas que atraviesan el silicio crean pares electrón-hueco que se desplazan en un campo eléctrico a través del silicio, obteniéndose una señal. Un detector de tiras de silicio de gran área, conocido como TT, tiene cuatro capas de detección y está situado aguas arriba del [imán dipolar](http://lhcb-public.web.cern.ch/lhcb-public/en/detector/Magnet-en.html). Las partículas cargadas son desviadas por la fuerza magnética, y el imán dipolar tiene un poder de curvatura de aproximadamente 4 Tm. La dirección del campo magnético se invierte regularmente. Esto permite controlar las incertidumbres sistemáticas debidas a que partículas con cargas opuestas se curvan hacia diferentes regiones del detector, con eficiencias de detección potencialmente diferentes. Tres estaciones de [detectores de tiras de silicio](http://lhcb-public.web.cern.ch/lhcb-public/en/detector/Trackers-en.html) (cada una con cuatro capas de detección) y tubos de deriva de paja (straw drift tubes) se colocan aguas abajo del imán y se conocen como T1, T2, T3. Los tubos de deriva de paja están llenos de un gas (Ar y CO2) que es ionizado por las partículas cargadas que lo atraviesan, y esta carga se desplaza bajo un campo eléctrico para ser medida. 

Los puntos medidos se utilizan para reconstruir la trayectoria de las partículas cargadas. El momento de la partícula cargada se puede obtener a partir de la curvatura de la trayectoria. El sistema de seguimiento combinado proporciona una medición del momento con una incertidumbre relativa que varía del 0.5% a bajo momento (p) al 1% a 200 GeV/c, y una medición del parámetro de impacto con una resolución de 20 μm para partículas cargadas con gran momento transverso, p<sup>T</sup>. 

Diferentes tipos de hadrones cargados (piones, kaones y protones) se distinguen utilizando información de dos [detectores Cherenkov de anillado](http://lhcb-public.web.cern.ch/lhcb-public/en/detector/RICH-en.html). Las partículas cargadas viajan a través de volúmenes de gases (CF<sub>4</sub>, C<sub>4</sub>F<sub>10</sub>) o sólidos ultraligeros (aerogel) más rápido que la velocidad de fase local de la luz en esos materiales y emiten fotones. Esto se conoce como efecto Cherenkov. Estos fotones se miden y el ángulo al que se emite la luz está directamente relacionado con la velocidad de la partícula. Combinando la información de la velocidad con la información del momento, se puede deducir la masa de la partícula y utilizarla para determinar el tipo de hadrón cargado. 

Los candidatos a fotones, electrones y hadrones son identificados por un sistema de calorímetros que consiste en detectores de almohadilla centelleadora (SPD) y pre-calentador (PS), un calorímetro electromagnético (ECAL) y un calorímetro hadrónico (HCAL). Los [calorímetros](http://lhcb-public.web.cern.ch/lhcb-public/en/detector/Calorimeters-en.html) se basan en capas alternas de hierro y centelleador. Las partículas interactúan y son absorbidas en el hierro, y la cantidad de luz de centelleo detectada permite reconstruir la energía de las partículas. Los fotones y electrones son absorbidos principalmente en el calorímetro electromagnético. Los hadrones son absorbidos principalmente en el calorímetro hadrónico. Solo los neutrinos, que permanecen sin detectar, y los muones típicamente atraviesan los calorímetros. Los muones son identificados por un [sistema de muones](http://lhcb-public.web.cern.ch/lhcb-public/en/detector/Muon-en.html) (M1–5) compuesto por capas alternas de hierro y cámaras proporcionales de múltiples hilos.

Aproximadamente 11 MHz de cruces de paquetes contienen una o más interacciones pp inelásticas en la muestra de datos estudiada aquí. El tamaño de los datos de evento producidos por el detector LHCb significa que esto solo podía escribirse en disco a una tasa de 3 kHz. El sistema utilizado para determinar qué eventos almacenar se conoce como [trigger (desencadenador)](http://lhcb-public.web.cern.ch/lhcb-public/en/Data%20Collection/Triggers2-en.html). Consiste en una etapa de hardware, basada en información de los sistemas de calorímetros y muones, seguida de una etapa de software, que aplica una reconstrucción completa del evento. La etapa de software se ejecuta en una gran granja de CPUs multiprocesador que ejecutan 26,110 copias del programa. El trigger tiene como objetivo mantener la mayor eficiencia posible para los canales de desintegración que se estudiarán, mientras rechaza eventos de bajo interés. 

Luego, los datos se procesan en el CERN y en computadoras de un sistema informático mundial conocido como GRID. Los datos se someten a una preselección para registrar eventos de interés para procesos físicos particulares. Grupos individuales de analistas escriben programas, utilizando el marco de software de LHCb, para realizar selecciones adicionales y reducir el tamaño de los datos, almacenando solo las variables de interés principal. Los datos que se te proporcionan en el experimento tienen esta forma. 

Las simulaciones, en las que se genera el proceso físico en estudio o los procesos de fondo, también se realizan para compararlas con los datos. Estas también pueden incluir simulaciones detalladas de la respuesta del detector, de modo que los datos de simulación sean directamente comparables con los datos. Aquí se proporcionan datos de simulación simplificados, que no incluyen la respuesta del detector. 

La muestra de datos proporcionada ha sido seleccionada por el trigger y la selección offline de aproximadamente 10<sup>14</sup> cruces de haces protón-protón que ocurrieron en LHCb en 2011. La muestra proporcionada incluye solo las variables reconstruidas más importantes necesarias para realizar este análisis, con el fin de reducir el tamaño del archivo de datos y disminuir el tiempo de procesamiento para tu análisis. Las muestras contienen 3.4 millones (5.1 millones) de eventos correspondientes a una [luminosidad](https://en.wikipedia.org/wiki/Luminosity_scattering_theory) integrada de 434 pb<sup>−1</sup> (584 pb<sup>−1</sup>) con el imán en la polaridad "hacia arriba" ("hacia abajo").
