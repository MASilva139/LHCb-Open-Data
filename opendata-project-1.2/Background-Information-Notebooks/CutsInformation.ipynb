# Cortes de Selección de Datos

## ¿Por qué aplicamos cortes?

En el LHC ocurre un gran número de colisiones protón-protón; en 2011, los haces se cruzaban aproximadamente 20 millones de veces por segundo durante las operaciones. Las capacidades de procesamiento y los requisitos de almacenamiento significan que solo podemos guardar unos pocos miles de estos eventos por segundo. La selección de qué eventos almacenar la realiza el sistema de [trigger](http://lhcb-public.web.cern.ch/lhcb-public/en/Data%20Collection/Triggers-en.html), que busca los eventos más interesantes para guardarlos.

Los procesos que queremos estudiar, como el de este análisis, son comparativamente raros. Los eventos debidos al proceso que deseamos estudiar se conocen como señal. Otros procesos que pueden falsificar la señal se conocen como fondos. Se aplican cortes de selección en el sistema de trigger y luego en el procesamiento offline para reducir los datos. Los cortes de selección se realizan sobre las características de los eventos para maximizar el número de eventos de señal seleccionados y minimizar el número de eventos de fondo.

Los principales cortes de selección que se han aplicado para seleccionar la muestra de datos que tienes se describen [aquí](DetectorSoftwareDataSample.ipynb). El tamaño total de la muestra de datos reales que tienes es de unos pocos millones de eventos.

## ¿Cómo elegimos los cortes ideales?

No es posible separar completamente los eventos de señal y de fondo. Si aplicamos una selección muy estricta, podemos reducir el número de eventos de fondo a un nivel bajo, pero esto también reducirá el número de eventos de señal seleccionados. Esto reducirá la sensibilidad estadística de tu medición. Una selección muy laxa mantendrá tantos eventos de señal como sea posible, pero también tendrá un gran número de eventos de fondo en tu muestra. Esto puede causar sesgos en tu medición, dependiendo de cómo trates el fondo, y generará incertidumbres debidas a las fluctuaciones estadísticas de la muestra de fondo.

In practice for this project you should simply choose a set of kaon identification cuts that produce a large signal peak in the invariant mass distribution over a comparatively small background.

En la práctica, para este proyecto simplemente debes elegir un conjunto de cortes de identificación de kaones que produzca un pico de señal grande en la distribución de masa invariante sobre un fondo comparativamente pequeño.

Aquí discutimos cómo se pueden optimizar los cortes. Una medida común que se utiliza para optimizar los cortes de selección aplicados es la determinación de la significancia de la señal. Esta medida es
<span style="white-space: nowrap; font-size:larger">
s / &radic;<span style="text-decoration:overline;">&nbsp;s + b&nbsp;</span>
</span>, donde **s** es el número estimado de eventos de señal y b es el número estimado de eventos de fondo. El numerador son los eventos de señal. El denominador es la incertidumbre estadística sobre el número de eventos observados, señal y fondo. Por lo tanto, representa una medida de la significancia de observar la señal sobre el fondo más la señal. Los cortes se pueden optimizar para que esta medida tenga el valor más alto posible.
