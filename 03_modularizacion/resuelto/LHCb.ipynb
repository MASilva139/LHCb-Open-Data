# Análisis de canales (Versión 01)

```python
from lhcb_analisis.config import (mK, mPi)
from lhcb_analisis.channels import run_channel_analysis
```

# Canal $B^{\pm} \to K^{\pm}K^{-}K^{+}$

```python
PRESELECTION_KKK = (
    '(H1_ProbK > 0.5) & '
    '(H2_ProbK > 0.5) & '
    '(H3_ProbK > 0.5) & '
    '(H1_ProbPi < 0.5) & '
    '(H2_ProbPi < 0.5) & '
    '(H3_ProbPi < 0.5) & '
    '(H1_isMuon == 0) & '
    '(H2_isMuon == 0) & '
    '(H3_isMuon == 0)'
)
MASSES_KKK = (mK, mK, mK)

results_KKK = run_channel_analysis(
    channel_name = "B±⇾K±K+K-",
    preselection = PRESELECTION_KKK,
    masses = MASSES_KKK,
    fit_model = 'gauss_exp',
    charm_veto = True,
    use_sideband = False,
    mass_window = (5194, 5364)
)
```
